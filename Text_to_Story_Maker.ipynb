```python
from transformers import pipeline
```

```python
story_generator = pipeline("text-generation", model="gpt2")
```

```python
def generate_story(prompt, max_length, num_return_sequences, temperature, top_k=50):

    print("Generating story...")
    stories = story_generator(
        prompt,
        max_length=max_length,
        num_return_sequences=num_return_sequences,
        truncation=True,
        pad_token_id=50256,
        do_sample=True,
        top_k=top_k,
        temperature=temperature
    )
    return [story['generated_text'] for story in stories]
```

```python
print("Welcome to the Text to Story Maker!")
user_prompt = input("Enter a sentence to start the story: ")

max_length = int(input("Enter the maximum length for the story (e.g., 150): "))
num_return_sequences = int(input("Enter the number of stories to generate (e.g., 1): "))
temperature = float(input("Enter the creativity level (temperature, e.g., 0.7): "))
top_k = int(input("Enter the top-k value for sampling diversity (e.g., 50): "))

generated_stories = generate_story(user_prompt, max_length, num_return_sequences, temperature, top_k)
print("\nHere is your story:\n")
for idx, story in enumerate(generated_stories, start=1):
     print(f"Story {idx}:")
     print(story)
     print("-" * 50)
```
